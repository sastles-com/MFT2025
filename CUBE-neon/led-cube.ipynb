```python
import pandas as pd
import numpy as np
import json
import math

import plotly.express as px
import plotly.graph_objects as go
from itertools import  starmap
```

```python
num = 64
panel = 6
pitch = 8
size = 65
```

```python
width  = int(math.sqrt(num))
height = int(math.sqrt(num))
print('width:', width, 'height:', height)
```

```python
plane = [
        [0,0,1,0], 
        [-1,0,0,3],
        [0,-1,0,1],
        [1,0,0,1],
        [0,1,0,3],
        [0,0,-1,2]]
print(plane)
```

```python
start = int(pitch/2+(int(width/2)-1)*pitch)
print('start', start)
```

```python
#                 https://www.mathpython.com/numpy-rotate-vector

def affine(param, vec):
    x = param[0]
    y = param[1]
    z = param[2]
    r = param[3]
    
    for rr in range(r):
        vec = [vec[1], -vec[0], vec[2]]

    if z != 0:
#         vec[2] = int(z*vec[2] + size/2)
        if z > 0:
            vec = [vec[0], vec[1], int(z*(vec[2] + size/2))]
        else:
            vec = [vec[0], -vec[1], int(z*(vec[2] + size/2))]
    elif x != 0:
        if x > 0:
            vec = [int(x*(vec[2] + size/2)), vec[1], -vec[0]]
        else:
            vec = [int(x*(vec[2] + size/2)), vec[1], vec[0]]
            
    elif y != 0:
        if y > 0:
            vec = [vec[0], int(y*(vec[2] + size/2)), -vec[1]]
        else:
            vec = [vec[0], int(y*(vec[2] + size/2)), vec[1]]

    
#     print(x,y,z,r, f"({vec[0]}, {vec[1]}, {vec[2]})")
    print(x,y,z,r, vec)
    return vec
```

```python
# df = DataFrame()
df = pd.DataFrame(index=[], columns=['x', 'y', 'z', 'id', 'num'])

for id, p in enumerate(plane):
    print(p)

    for y in range(height):
        yy = -start + y * pitch
        # print(yy)
        for x in range(width):
            xx = start - x * pitch
            panel_num = x + y * height

            vec = affine(p, [xx, yy, 0])
            position = pd.Series([vec[0], vec[1], vec[2],  id, panel_num], index=df.columns)
            df = df.append(position, ignore_index=True)

                
```

```python
# プロットを作成
plot = []
# for name, y in df.items():
for p in range(panel):
    pp = df[df['id'] == p]
    print(pp)
    d = go.Scatter3d(x=pp['x'],y=pp['y'],z=pp['z'],mode='markers',marker=dict(
        size=5
        ,color = pp['num']
#         ,color = pp.index
        ,colorscale = 'Viridis'
#         ,colorscale = 'Oranges'
    #     ,symbol= result_symbol
    ))
    plot.append(d)  # プロットを格納
```
