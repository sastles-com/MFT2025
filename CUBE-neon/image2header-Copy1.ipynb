```python
import pandas as pd
import numpy as np
import json
import math

import plotly.express as px
import plotly.graph_objects as go
from itertools import  starmap

import cv2
```

```python
width = 160


img =  cv2.imread("resize.jpg", cv2.IMREAD_GRAYSCALE)
h, w = img.shape[:2]
height = round(h * (width / w))
dst = cv2.resize(img, dsize=(width, height))

# img
width, height
dst
```

```python
filename = 'half-grad-' + str(width) + '-' + str(height) + '.h'
filename
```

```python


with open(filename, mode = "w", encoding="utf_8") as f:
    mes = "#define MASK_WIDTH " + str(width) + "\n"
    f.write(mes)
    mes = "#define MASK_HEIGHT " + str(height) + "\n"
    f.write(mes)
    f.write("#define TOTAL_MASK_PIXELS MASK_WIDTH*MASK_HEIGHT\n\n")
    f.write("int mask[MASK_HEIGHT][MASK_WIDTH] = { \n")

    for y in range(height):
        f.write(f"\t{{")
        for x in range(width):
            pos = x + y * width
#             print(pos, x, y, f"{df['red'][pos]}, ")
            print(pos, x, y, f"{dst[y, x]}, ")
            f.write(f"{dst[y, x]}, ")
#     for index, data in df.iterrows():
#         f.write(f"\t{{{data[0]}, {data[1]}, {data[2]}}}, \n")
        f.write("},\n")

    f.write("};\n")
 
```
